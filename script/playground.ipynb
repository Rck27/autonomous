```python
pole1 = (-80, 20)
pole2 = (-230, 20)

a = (-80, 100)
b = (220, 65)

ab = np.array(b) - np.array(a)

print(ab)

import numpy as np

p = np.array([pole1, pole2])

for i in p:
    ap = np.array(i) - np.array(a)
    c = np.cross(ab, ap)
    dot = np.dot(ab, ap)
    dist = np.abs(c)/ np.linalg.norm(ab)
    t = (np.dot(ab, ap)) / np.dot(ab, ab)
    print(f"pole {i}, cross {c}, dot {dot}, dist {dist}, t {t}")
```
